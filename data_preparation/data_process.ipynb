# Step 1. Get the dedeplicated list of ids of materials with magnetic atoms （Result: 89902）

```python
res_dic = {}

for atom in tqdm(magnetic_atoms):
    with MPRester(my_api_key) as mpr:
        docs = mpr.materials.summary.search(elements=[atom], 
                                  fields=["material_id", "band_gap"])
        mpid_bgap_dict = {doc.material_id: doc.band_gap for doc in docs}
        res_dic.update({atom: [str(i) for i in mpid_bgap_dict]})
```

```python
mp_id_set_list = [v for k,v in res_dic.items()]
```

```python
merged_list = []
for sublist in mp_id_set_list:
    merged_list.extend(sublist)
```

```python
# Identify duplicates (Wu:This step is unnecessary, it is just for me to understand things.)

# Wu: Why there is duplicate MP IDs? 
# Answer: for example, mp-1174608, which is Li4MnCo2O7, containing two magnetic elements: Mn and Co. We use Mn to find the ID, we find once; 
# we use Co to find, again, we find this material, this is twice, so duplicate.

# Explanation:
# merged_list = [1, 2, 3, 4, 2, 5, 3, 6]
# duplicate_list = find_duplicates(merged_list)
# print(duplicate_list)  # Output: [2, 3]

def find_duplicates(input_list):
    seen = set()
    duplicates = set()
    for item in input_list:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    return list(duplicates)

duplicate_list = find_duplicates(merged_list)
print(duplicate_list)
```

```python
# Remove duplicates
deduplicated_list = list(set(merged_list)) 
```

```python
# Play with the full list: deduplicated_list
# The list is too long to work, use the next cell for Batch Processing.
with MPRester(my_api_key) as mpr:
    docs = mpr.materials.summary.search(material_ids=deduplicated_list, fields=["material_id","composition", "symmetry","structure"])
```

```python
# Use batch processing to work with the long list deduplicated_list.
# Define a batch size, the number of materials you work for one time
batch_size = 10000

# Function to fetch data in batches
def fetch_data_in_batches(ids, batch_size):
    batches = [ids[i:i + batch_size] for i in range(0, len(ids), batch_size)]
    all_results = []
    with MPRester(my_api_key) as mpr:
        for batch in batches:
            results = mpr.materials.summary.search(material_ids=batch, fields=["material_id","composition", "symmetry","structure"])
            all_results.extend(results)
    return all_results

# Fetch data for all IDs in batches
docs = fetch_data_in_batches(deduplicated_list, batch_size)
```

# Step 2. Remove materials with two or more types of magnetic atoms (66979) & Split up materials with odd (18090) or even (48889) number of magnetic atoms

```python
single_type_atom_materials = {}             
odd_atoms_materials = {}
even_atoms_materials = {}

for doc in docs:
    my_string = str(doc.composition)

    # Define a regular expression pattern to match elements and numbers
    pattern = re.compile(r"([A-Z][a-z]*)(\d*)")

    # Iterate over matches and extract element and number
    ma_set = []
    num_set = []
    
    for match in pattern.finditer(my_string):
        element = match.group(1)
        number_str = match.group(2)
        if number_str:
            number = int(number_str)
        else:
            raise NotImplementedError
        
        if element in magnetic_atoms:
            ma_set.append(element)
            num_set.append(number)
    if len(ma_set) == 1:
        single_type_atom_materials.update({str(doc.material_id):doc})
        if num_set[0] % 2 == 1:
            odd_atoms_materials.update({str(doc.material_id):doc})

for k,v in single_type_atom_materials.items():
    if k not in odd_atoms_materials:
        even_atoms_materials.update({k:v})
```

# Step 3: Find negative materials among materials with even number of magnetic atoms

## Condition1: with space-inversion symmetry (31073)

```python
condition_1 = list(range(47,75)) + list(range(83,89)) + list(range(123,143)) + list(range(147,149)) + list(range(162,168)) \
                + list(range(175,177)) + list(range(191,195)) + list(range(200,207)) + list(range(221,231)) \
                + [2, 10, 11, 12, 13, 14, 15]
con1_result = {}

# 判断symmetry number是否在列表里，提取在里面的
# # symmetry.number is the international number for the space group
for k,v in even_atoms_materials.items():
    if v.symmetry.number in condition_1:
        con1_result.update({k:v}) 

print(f"The number of materials with inversion center among the even_atoms_materials is : {len(con1_result)}")
```

## Condition2: with two magnetic atoms in primitive unit cell (9903)

```python
# 在condition1的结果中，判断这个Cr元素后面的数字，提取是2的
# Based on con1_result
con2_result = {}
for k,doc in con1_result.items():
    my_string = str(doc.composition)

    # Define a regular expression pattern to match elements and numbers
    pattern = re.compile(r"([A-Z][a-z]*)(\d*)")

    # Iterate over matches and extract element and number
    ma_set = []
    num_set = []
    
    for match in pattern.finditer(my_string):
        element = match.group(1)
        number_str = match.group(2)
        if number_str:
            number = int(number_str)
        else:
            raise NotImplementedError
        
        if element in magnetic_atoms:
            ma_set.append(element)
            num_set.append(number)
    if len(ma_set) == 1:
        single_type_atom_materials.update({str(doc.material_id):doc})
        if num_set[0] == 2:
            con2_result.update({k:doc})
            
print(f"In the result of Condition 1, the number of materials with two magnetic atoms in primitive cell is : {len(con2_result)}") 
```

## Condition3: magnetic atoms both are at the inversion centers (for the purpose of identifying negative samples of altermagnets) (3602)

#### Note: thres=1e-5, in the original Gao's paper, it is 0.01. If it is 0.01, get 3041, if it is 1e-1, get 2964.
#### Now, I add Body diagonal midpoints to the inversion center list, I get 3602.

```python
# 在condition2结果中，提取出至少有一个不在这个里面的     # 难道说至少有一个原子不在中心对称点吗？
# Based on con_result
# condition_3 has 8 conditions. See below, each row is a condition.
# Wu: If the coordinates of the given list (see below, the site coordinates) match any one of the 8 conditions in condition_3, 
# the match_condition_3 function will return True; 
# if the given list matches none of the 8 conditions, the match_condition_3 function will return False.
# thres=0.01 by authors, isn't too big? maybe 1e-5
# Can use break: Once a non-matching site is found, break to avoid redundant checks. 

# List of potential inversion centers in the primitive unit cell
condition_3 = [
[0.0000, 0.0000, 0.0000],
[0.0000, 0.0000, 0.5000],
[0.0000, 0.5000, 0.0000],
[0.0000, 0.5000, 0.5000],
[0.5000, 0.0000, 0.0000],
[0.5000, 0.0000, 0.5000],
[0.5000, 0.5000, 0.0000],
[0.5000, 0.5000, 0.5000],
[0.2500, 0.2500, 0.2500], # Body diagonal midpoints
[0.7500, 0.7500, 0.7500], # Body diagonal midpoints
# Add more positions if needed based on the specific symmetry
]

def match_condition_3(list, thres=1e-5):    
    for condition in condition_3:
        if (abs(condition[0] - list[0]) < thres) and (abs(condition[1] - list[1]) < thres) and (abs(condition[2] - list[2]) < thres):
            return True
    return False

# If an entry has any site with coordinates that do not match any condition within a given threshold, it is added to con3_result_pre.
con3_result_pre = {}
for k,doc in con2_result.items():
    for site in doc.structure.sites:
        if not site.specie.as_dict()['element'] in magnetic_atoms:
            continue
        docsite = [site.a, site.b, site.c]
        if not match_condition_3(docsite):    
            con3_result_pre.update({k:doc})

con3_result = {}
for k,v in con2_result.items():
    if k not in con3_result_pre:
        con3_result.update({k:v})

print(len(con3_result_pre))
print(len(con3_result)) 
```

# Step 6: make csv files for the IDs of the materials using txt files 

```python
# reads text files, extracts specific patterns using regular expressions, 
# and then writes the extracted data into CSV files using the pandas library.
# re is used for regular expression operations.

import os
import re
import pandas as pd

# Relative path, in the same folder
base_dir = 'save_files/'

candidate_pre_file = os.path.join(base_dir, "candidate_pre.txt")
negative_pre_file = os.path.join(base_dir, "negative_pre.txt")

# candidate_pre_file
with open(candidate_pre_file, 'r') as file:
    file_content = file.read()
pattern = r"'(mp-\d+)'" 
# A regular expression pattern is defined to match strings of the form "mp-<digits>".
# The parentheses () around mp-\d+ form a capturing group,
# which allows you to extract the portion of the text 
# that matches this part of the pattern (excluding the surrounding quotes).

# Find all matches
matches = re.findall(pattern, file_content)                      # list

candidate_pre_df = pd.DataFrame(matches, columns=["ID"])             # DataFrame
candidate_pre_df.to_csv(os.path.join(base_dir, 'candidate_pre.csv'), index=False)

# negative_pre_file
with open(negative_pre_file, 'r') as file:
    file_content = file.read()
pattern = r"'(mp-\d+)'"

# Find all matches
matches = re.findall(pattern, file_content)

negative_df = pd.DataFrame(matches, columns=["ID"])
negative_df.to_csv(os.path.join(base_dir, 'negative_pre.csv'), index=False)
```

```python
# Find mvc-11882
def find_missing_entries(a, b):
    """
    Finds the entries in list 'a' that are not present in list 'b'.

    Args:
        a: The first list.
        b: The second list.

    Returns:
        A new list containing the elements found in 'a' but not in 'b'.
    """
    missing_entries = []
    for item in a:
        if item not in b:
            missing_entries.append(item)
    return missing_entries

# Example usage
a = candidate_pre_list
b = candidate_pre_df_list

missing_entries = find_missing_entries(a, b)

print("Elements in list 'a' not present in list 'b':", missing_entries) 
```
